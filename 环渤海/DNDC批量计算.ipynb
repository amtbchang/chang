```python
#####1
import os
import time

name = ['mfp','rmn0.9','rmn0.8','rmn0.7','rmn0.6','rmn0.5','linghai_FP',
        'linghai_FP_0.9','linghai_FP_0.8','linghai_FP_0.7','linghai_FP_0.6',
        'linghai_FP_0.5','straw0.2','straw0.4','straw0.6','straw0.8','straw']
dadc_path = r'C:\bin\linghai\gi_85'

os.chdir(path)


kk = []
for i in name:
    kk.append('dndc.exe '+ dadc_path + os.sep + i + '.dnd')

###
from multiprocessing.pool import ThreadPool
from concurrent.futures import ThreadPoolExecutor

 
def dndc_run(filepath):
    t1 = time.time()
    os.system(filepath)
    print('%.3f'(time.time()-t1))

begin = time.time()
executor=ThreadPoolExecutor(4)
executor.map(dndc_run, kk)
end = time.time()
print ('Run_time::',end-begin)
```
